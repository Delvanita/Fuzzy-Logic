```python
#membaca file

d = p.read_csv('mahasiswa.csv', sep=';')
data = d.values.tolist()
data
```

```python
hasilan = []
for i in range(len(data)):
    tmp = float(data[i][1])
    hasilan.append(tmp)
print(hasilan)

pengluaran = []
for i in range(len(data)):
    pengluaran.append(data[i][2])
print(pengluaran)
```

```python
#fungsi keanggotaan penghasilan

#PenghasilanRendah

def PenghasilanRendah(peng):
    if((peng > 4.0) and (peng <= 6.0 )):
        return (6.0 - peng )/(6.0 - 4.0)
    elif (peng <= 4.0):
        return 1
    elif (peng > 6.0):
        return 0
print(PenghasilanRendah(8))



#PenghasilanSedang

def PenghasilanSedang(peng):
    if((peng > 4.0) and (peng < 6.0)):
        return (peng - 4.0)/(6.0 - 4.0)
    elif((peng > 13.0) and (peng <= 17.0)):
        return (17.0 - peng)/(17.0 - 13.0)
    elif ((peng <= 4.0) or (peng > 17.0)):
        return 0
    elif ((peng >= 6.0) and (peng <= 13.0)):
        return 1
print(PenghasilanSedang(10))


#Penghasilan Tinggi

def PenghasilanTinggi(pengT):
    if((pengT <= 17.0) and (pengT > 13.0)):
        return (pengT - 13.0)/(17.0 - 13.0)
    elif (pengT <= 13.0):
        return 0
    elif (pengT > 17.0):
        return 1
print(PenghasilanTinggi(19))
```

```python
#fungsi keanggotaan pengeluaran


def PengeluaranRendah(r):
    if((r > 4.0) and (r <= 6.0)):
        return (1 - r )/(6.0 - 4.0)
    elif (r <= 4.0):
        return 1
    elif (r > 6.0):
        return 0
print(PengeluaranRendah(7))


#PengeluaranSedang

def PengeluaranSedang(s):
    if((s > 4.0) and (s < 6.0)):
        return (s - 4.0)/(6.0 - 4.0)
    elif((s > 8.0) and (s < 10.0)):
        return (10.0 - s)/(10.0 - 8.0)
    elif ((s <= 4.0) or (s >= 10.0)):
        return 0
    elif ((s >= 6.0) and (s <= 8.0)):
        return 1
print(PengeluaranSedang(6))


#PengeluaranTinggi

def PengeluaranTinggi(t):
    if((t >= 8.0) and (t < 11.0)):
        return (t - 8.0)/(11.0 - 8.0)
    elif (t <= 8.0):
        return 0
    elif (t >= 11.0):
        return 1
print(PengeluaranTinggi(4))
```

```python
#l : low
#m : mid
#b : bottom

def AcceptRules(l, m, b ):
    ar1 = (min(l, m))
    ar2 = (min(l, b))
    values = max(ar1, ar2)
    return values
```

```python
#high
#average
#low
#upper
#midle
#bottom

def considerRules(lo, av, up, mid, bot):
    cons1 = (min(lo, up))
    cons2 = (min(av, up))
    cons3 = (min(av, mid))
    cons4 = (min(av, bot))
    values = max(cons1, cons2, cons3, cons4)
    return values
```

```python
#low
#upper
#midle

def rejectRules(hi, up, mid, bot):
    reject1 = (min(hi, up))
    reject2 = (min(hi, mid))
    reject3 = (min(hi, bot))
    values = max(reject1, reject2, reject3)
    return values
```

```python
#metode sugeno

def defuzzification (a, c, r, i):
    acc = 100
    cons = 75
    rej = 25
    return ((acc*a)+(cons*c)+(rej*r))/(a + c + r), i
```

```python


def fuzzificationpenghasilan(x):
    values = []
    values.append(PenghasilanRendah(x))
    values.append(PenghasilanSedang(x)) 
    values.append(PenghasilanTinggi(x))
    return values
```

```python
def fuzzificationpengeluaran(x):
    values = []
    values.append(PengeluaranRendah(x))
    values.append(PengeluaranSedang(x))
    values.append(PengeluaranTinggi(x))
    return values
```

```python
#main

has= [] 
pengl = []
score = []
idx = []
for i in range (100):
    has.append(fuzzificationpenghasilan(hasilan[i]))
    pengl.append(fuzzificationpengeluaran(pengluaran[i]))

for i in range (100):
#     penghasilan
    high = has[i][0]
    average = has[i][1]
    low = has[i][2]
#     pengeluaran
    upper = pengl[i][0]
    midle = pengl[i][1]
    bottom = pengl[i][2]
    
    acc = AcceptRules(low, midle, bottom)
    cons = considerRules(low, average, upper, midle, bottom)
    rej = rejectRules(high, upper, midle, bottom)
    
    score.append(defuzzification(acc, cons, rej, i))
score.sort(reverse = True)
print(score)
```

```python
#ouput

Fix = []
for i in range (20):
    Fix.append(score[i])
#     print(i+1, hasilan[Fix[i][1]], pengluaran[Fix[i][1]] ,Fix[i][0])
Id = []
for i in range(20):
    Id.append(Fix[i][1])
save = p.DataFrame(Id)
save
```
